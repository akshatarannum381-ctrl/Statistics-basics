Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.


-> Descriptive statistics are used to describe, summarize, or present data in a meaningful way. They do not make predictions or generalizations beyond the data you have. Example:
Suppose you surveyed 50 students and found their average height is 165 cm.
That’s descriptive statistics because you are only describing those 50 students.
Whereas in inferential statistics are used to make conclusions, predictions, or generalizations about a larger population based on a sample of data. Example:
You take a sample of 50 students from a university of 10,000 students and find their average height is 165 cm.
You then infer that the average height of all university students is around 165 cm.
That’s inferential statistics because you are generalizing from a sample to a population.

Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.

-> Sampling is the process of selecting a small group (sample) from a larger group (population) to study and draw conclusions about the entire population.
Since it’s often impossible or expensive to study an entire population, we collect data from a sample that represents the whole group.

Example:
If a school has 1,000 students, and you select 100 students to find the average height, those 100 students are your sample, and 1,000 students are the population.

Question 3: Define mean, median, and mode. Explain why these measures of central

-> Measures of central tendency are statistical tools that help identify the center or typical value in a dataset.
The three main measures are Mean, Median, and Mode.
The mean is the sum of all observations divided by the total number of observations.

The median is the middle value when the data is arranged in ascending or descending order.

Example:
For the marks: 50, 60, 70, 80, 90
→ The middle value is 70, so the median = 70.

If there are even numbers of observations, the median is the average of the two middle values.
The mode is the value that occurs most frequently in the dataset.

Example:
In 5, 7, 8, 8, 10 → the number 8 occurs twice,
so the mode = 8.

If all numbers occur only once, there is no mode.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?

-> Skewness measures the asymmetry (unevenness) of a data distribution around its mean.
It tells us which side of the mean (left or right) has a longer tail. Whereas kurtosis measures the peakedness or flatness of a data distribution compared to a normal (bell-shaped) curve.
It shows how concentrated or spread out the data is around the mean.
A positive skew means that most data values are low, and a few high values stretch the tail to the right. Example:
Income distribution — most people earn average or below-average income, but a few earn very high salaries.

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

cov_matrix = np.cov(list_x, list_y, bias=False)
covariance = cov_matrix[0, 1]

correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

Output:
Covariance: 225.0
Correlation Coefficient: 0.9933992677987826
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

sns.boxplot(data=data)
plt.title("Boxplot of Given Data")
plt.xlabel("Values")
plt.show()

A boxplot is a graphical summary of data based on five key statistics:

Minimum (excluding outliers)

First Quartile (Q1) – 25% of data below this value

Median (Q2) – Middle value of the data

Third Quartile (Q3) – 75% of data below this value

Maximum (excluding outliers)

Interquartile Range (IQR) = Q3 – Q1
Outliers are identified as:

Values below Q1 – 1.5 × IQR

Values above Q3 + 1.5 × IQR
Outlier: 35 (since it is greater than 31.75)

All other values are within the normal range.
```

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

cov_matrix = np.cov(advertising_spend, daily_sales, bias=False)
covariance = cov_matrix[0, 1]

correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

Output:
Covariance: 87500.0
Correlation Coefficient: 0.9912407071619309

Covariance = 87500.0 → Positive value indicates that when advertising spending increases, sales also increase.

Correlation ≈ 0.99 → Very strong positive linear relationship.
```

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt
import statistics as stats

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

mean_value = stats.mean(survey_scores)
median_value = stats.median(survey_scores)
mode_value = stats.mode(survey_scores)
std_dev = stats.stdev(survey_scores)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Standard Deviation:", std_dev)

plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Distribution")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.show()

Output:
Mean: 7.47
Median: 7
Mode: 7
Standard Deviation: 1.67

| Measure                     | Purpose                    | What It Tells You                                            |
| --------------------------- | -------------------------- | ------------------------------------------------------------ |
| **Mean**                    | Average satisfaction level | Overall satisfaction trend                                   |
| **Median**                  | Middle value               | Typical satisfaction score (less affected by outliers)       |
| **Mode**                    | Most common score          | Most frequently given rating                                 |
| **Standard Deviation (SD)** | Measures spread of data    | High SD → opinions vary widely; Low SD → consistent opinions |
| **Minimum & Maximum**       | Range of scores            | Lowest and highest satisfaction                              |

| Visualization            | Purpose                                    | Description                                     |
| ------------------------ | ------------------------------------------ | ----------------------------------------------- |
| **Histogram**            | Shows frequency distribution               | Helps you see how scores are spread across 1–10 |
| **Boxplot**              | Highlights median, quartiles, and outliers | Good for quick summary                          |
| **Bar Chart (for mode)** | Shows most common score                    | Helpful for categorical frequency view          |


```
